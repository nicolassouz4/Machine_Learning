## Machine Learning Section - K Means


### Training and Testing Data

```python
# Using KMeans from sklearn.cluster
from sklearn.cluster import KMeans
```

```python
# Creating an instance of a KMeans() model named kmeans with 2 clusters
kmeans = KMeans(n_clusters=2)
```

```python
# Train/fit kmeans on the training data (except for the Private label)
kmeans.fit(college.drop('Private', axis=1))
```

```python
# Cluster center vectors
kmeans.cluster_centers_
```

### "Evaluation" (just because I have the right answer)

There is no perfect way to evaluate clustering if you don't have the labels, however as this is just an exercise we do have the labels so I took advantage of that to evaluate the clusters.

With real data we don't have that luxury hahah

```python
# Creating a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school
college['Cluster'] = college['Private'].map(lambda x: 1 if x == 'Yes' else 0)
```

```python
college.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(college['Cluster'], kmeans.labels_))
print("\n")
print(classification_report(college['Cluster'], kmeans.labels_))
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title("Original")
ax1.hist(college['Cluster'])
ax2.set_title('K Means')
ax2.hist(pd.DataFrame(kmeans.labels_))
```
